This is the script to convert Qchem input format xyz to standard xyz format as ASE
Line 1: The total number of atoms in the file (your dimer has exactly 74 atoms).
Line 2: A comment line (can be left blank or give a title).
Followed by: Clean atom labels and coordinates.


```python
import os
import glob

# Set the path to where your files are right now
directory_path = "../../15-Principle-structure/distorted-str"
txt_files = glob.glob(os.path.join(directory_path, "*.txt"))

# # Since the script is inside the folder, use "." for the current directory
# directory_path = "."

# # Find all matching text files locally
# txt_files = glob.glob(os.path.join(directory_path, "structure_*.txt"))

print(f"DEBUG: Found {len(txt_files)} total files in directory: '{os.path.abspath(directory_path)}'")

if len(txt_files) == 0:
    print("\n[WARNING]: Found 0 files. Make sure you are running the script from inside the distorted-str directory.")
else:
    for file_path in txt_files:
        with open(file_path, "r") as f:
            lines = f.readlines()
        
        clean_coords = []
        for line in lines:
            stripped = line.strip()
            if not stripped or stripped.startswith("$") or stripped == "1 2":
                continue
            clean_coords.append(line)
            
        num_atoms = len(clean_coords)
        
        new_xyz_path = os.path.splitext(file_path)[0] + ".xyz"
        with open(new_xyz_path, "w") as f:
            f.write(f"{num_atoms}\n")
            f.write("Converted from QChem format for SOAP analysis\n")
            f.writelines(clean_coords)

    print(f"\nSuccessfully converted {len(txt_files)} files to standard .xyz format!")
```
